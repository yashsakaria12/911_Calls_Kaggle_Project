# 911 Calls Capstone Project

For this capstone project we will be analyzing some 911 call data from [Kaggle](https://www.kaggle.com/mchirico/montcoalert). The data contains the following fields:

* lat : String variable, Latitude
* lng: String variable, Longitude
* desc: String variable, Description of the Emergency Call
* zip: String variable, Zipcode
* title: String variable, Title
* timeStamp: String variable, YYYY-MM-DD HH:MM:SS

* e: String variable, Dummy variable (always 1)

## Creating new features

** In the titles column there are "Reasons/Departments" specified before the title code. These are EMS, Fire, and Traffic. This is used to create a new column called "Reason" that contains this string value.** 

**For example, if the title column value is EMS: BACK PAINS/INJURY , the Reason column value would be EMS. **

```python
df['Reason'] = df['title'].apply(lambda x:x.split(':')[0])
df
```

** What is the most common Reason for a 911 call based off of this new column? **

```python
df['Reason'].value_counts()
```

**Countplot of 911 calls by Reason. **

```python
sns.countplot(df['Reason'])
```

___
**What is the data type of the objects in the timeStamp column? **

```python
type(df['timeStamp'][0])
```

**Convert the timeStamp column from strings to DateTime objects. **

```python
df['timeStamp'] = df['timeStamp'].apply(lambda x:pd.to_datetime(x))
```

```python
df['timeStamp'][0]
```

**Create 3 new columns called Hour, Month, and Day of Week using the timeStampcolumn.**

```python
df['Hour'] = df['timeStamp'].apply(lambda x:x.hour)
df['Month'] = df['timeStamp'].apply(lambda x:x.month)
df['Day'] = df['timeStamp'].apply(lambda x:x.dayofweek)
df
```

** Map the actual string names to the day of the week: **

```python
dmap={0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
df['Day'] = df['Day'].map(dmap)
df
```

** Countplot of the Day of Week column with the hue based off of the Reason column. **

```python
sns.countplot(df['Day'],hue=df['Reason']).legend(bbox_to_anchor=(1,0.7))
```

**Same plot repeated for month column**

```python
sns.countplot(df['Month'],hue=df['Reason']).legend(bbox_to_anchor=(1,0.7))
```

** Create a gropuby object called byMonth, where the DataFrame is grouped by the month column**

```python
mn = df.groupby('Month').count()
mn.head()
```

** Plot indicating the count of calls per month. **

```python
mn['lat'].plot()
```

** Create a linear fit on the number of calls per month. **

```python
mn['Month'] = mn.index
sns.lmplot(x = 'Month',y='twp',data=mn)
```

**Create a new column called 'Date' that contains the date from the timeStamp column. ** 

```python
df['Date'] = df['timeStamp'].apply(lambda x:x.date())
```

** Create a plot of counts of 911 calls.**

```python
mn = df.groupby('Date').count()
mn['lat'].plot()
```

** Recreate this plot but create 3 separate plots with each plot representing a Reason for the 911 call**

```python
r_traffic = df[df['Reason'] == 'Traffic']
r_traffic
byMonth = r_traffic.groupby('Date').count()
byMonth['lat'].plot()
```

```python
r_traffic = df[df['Reason'] == 'Fire']
r_traffic
byMonth = r_traffic.groupby('Date').count()
byMonth['lat'].plot()
```

```python
r_traffic = df[df['Reason'] == 'EMS']
r_traffic
byMonth = r_traffic.groupby('Date').count()
byMonth['lat'].plot()
```

** Heat maps**

```python
mat = df.groupby(['Day','Hour'])['Reason'].count().unstack()
mat
```

**Another way to create a matrix form for heatmaps is using the pivot table. This is how the pivot table object can be used to create the matrix table.**

```python
mat = df.pivot_table(index='Day',columns='Hour',aggfunc='count')
mat['lat']
```

** Now create a HeatMap using this new DataFrame. **

```python
plt.figure(figsize=(12,6))
sns.heatmap(mat,cmap='viridis')
```

** Now creating a clustermap using this DataFrame. **

```python
sns.clustermap(mat,cmap='viridis')
```
